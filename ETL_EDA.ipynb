```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
url = "https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-panaderia-panaderia-propia/_/N-1s5i92o"
page = urlopen(url)
html = page.read().decode("utf-8")
soup = BeautifulSoup(html, "html.parser")
```

```python
# Busca el elemento con la clase 'descrip_full' que tiene el mismo ID que el nombre del producto
producto = soup.find('div', id='descrip_full_sku00044409')

# Verifica si se encontró el elemento del producto
if producto:
    # Obtiene el siguiente elemento después del producto
    siguiente_elemento = producto.find_next_sibling()

    # Verifica si el siguiente elemento existe y extrae su texto (que sería el precio)
    if siguiente_elemento:
        precio = siguiente_elemento.get_text()
        print("Precio del producto:", precio)
    else:
        print("No se encontró el precio del producto.")
else:
    print("No se encontró el producto.")
```

```python
# Busca el elemento span con la clase 'atg_store_newPrice'
precio_span = soup.find('span', class_='atg_store_newPrice')

# Verifica si se encontró el elemento del precio
if precio_span:
    # Obtiene el texto dentro del elemento span (que debería ser el precio)
    precio = precio_span.get_text(strip=True)
    print("Precio del producto:", precio)
else:
    print("No se encontró el precio del producto.")
```

```python
import time
import re
import pandas as pd
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Nombre del archivo del controlador de Microsoft Edge (msedgedriver.exe)
EDGE_DRIVER_PATH = 'msedgedriver.exe'

# URL base de la página web con los productos
url_base = 'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-panaderia/_/N-s3bf1a'

# Configurar el servicio del controlador de Edge
edge_service = Service(EDGE_DRIVER_PATH)
edge_service.start()

# Configurar las opciones del navegador
edge_options = webdriver.EdgeOptions()
edge_options.headless = True  # Para ejecutar el navegador en modo sin cabeza

# Iniciar el navegador Edge
driver = webdriver.Edge(service=edge_service, options=edge_options)

# Función para obtener productos y precios de una página
def obtener_productos_precios(html_pagina):
    patron_producto_precio = r'(?:<div class="descrip_full" id="descrip_full_sku\d+" style="display: none;">(.*?)<\/div>.*?<span class="atg_store_newPrice">\s*([^\s]+)\s*<\/span>)'
    productos_precios_encontrados = re.findall(patron_producto_precio, html_pagina, re.DOTALL)
    productos = []
    precios = []
    for producto, precio in productos_precios_encontrados:
        productos.append(producto.strip())
        precios.append(precio.strip())
    return productos, precios

# Lista para almacenar todos los productos y precios
todos_productos = []
todos_precios = []

# Función para navegar a través de las páginas y obtener los datos
def obtener_datos_todas_paginas(url):
    driver.get(url)
    while True:
        # Esperar hasta que los productos estén cargados
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "descrip_full")))
        # Obtener el HTML completo de la página
        html_pagina = driver.page_source
        # Obtener productos y precios de la página actual
        productos, precios = obtener_productos_precios(html_pagina)
        # Agregar productos y precios a la lista general
        todos_productos.extend(productos)
        todos_precios.extend(precios)
        # Buscar el enlace de "siguiente página"
        try:
            next_page_link = driver.find_element_by_link_text('Sig')
            # Hacer clic en el enlace de "siguiente página"
            next_page_link.click()
            # Esperar un momento para que la página cargue
            time.sleep(3)
        except:
            # Si no hay más páginas, salir del bucle
            break

# Obtener datos de todas las páginas
obtener_datos_todas_paginas(url_base)

# Crear un DataFrame de pandas con todos los productos y precios
df_todos_productos = pd.DataFrame({'Producto': todos_productos, 'Precio': todos_precios})

# Imprimir el DataFrame
print("DataFrame con todos los productos y precios:")
print(df_todos_productos)

# Cerrar el navegador
driver.quit()
```
