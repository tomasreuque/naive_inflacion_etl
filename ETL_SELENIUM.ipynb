```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import WebDriverException
from selenium.common.exceptions import NoSuchElementException


def obtener_numero_productos(url):
    try:
        # Inicializar el navegador Edge sin especificar opciones
        EDGE_DRIVER_PATH = 'msedgedriver.exe'
        edge_service = Service(EDGE_DRIVER_PATH)
        edge_service.start()
        edge_options = webdriver.EdgeOptions()
        edge_options.headless = True
        
        # Iniciar el navegador Edge
        driver = webdriver.Edge(service=edge_service, options=edge_options)
        
        # Obtener la página web
        driver.get(url)
            
        # Esperar hasta que el elemento con la id 'resultsCount' esté presente en la página
        numero_productos_element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, "resultsCount"))
        )
        
        # Extraer el número de productos del texto del elemento
        numero_productos = int(numero_productos_element.text.strip())
        
        # Encontrar el enlace a la siguiente página
        enlace_pagina_siguiente = driver.find_element(By.CSS_SELECTOR, "a[title='Ir a página 2']").get_attribute("href")
        
        # Reemplazar el número de resultados por página en el enlace de la segunda página
        enlace_pagina_siguiente = enlace_pagina_siguiente.rsplit('&Nrpp=', 1)[0] + f'&Nrpp={numero_productos}'
        
        # Cerrar el navegador
        driver.quit()
        
        # Guardar la información en un DataFrame
        df_info = pd.DataFrame({'Numero_productos': [numero_productos], 'Enlace_pagina_siguiente': [enlace_pagina_siguiente]})
        
        # Guardar el DataFrame en un archivo CSV
        df_info.to_csv('info_productos.csv', index=False)
        
        # Llamar a la función para obtener los productos con precios en la segunda página
        obtener_productos_con_precios(enlace_pagina_siguiente)
        
    except WebDriverException as e:
        print("Error al iniciar el navegador:", e)

def obtener_productos_y_precios(url):
    try:
        # Iniciar el navegador Chrome (asegúrate de tener el controlador de Chrome adecuado)
        CHROME_DRIVER_PATH = 'chromedriver.exe'
        chrome_service = Service(CHROME_DRIVER_PATH)
        chrome_service.start()
        chrome_options = webdriver.ChromeOptions()
        chrome_options.headless = True
        driver = webdriver.Chrome(service=chrome_service, options=chrome_options)
        
        # Obtener la página web
        driver.get(url)
        
        # Aquí agregarías la lógica para obtener todos los registros de productos con sus nombres y precios
        # Supongamos que obtienes una lista de productos con sus nombres y precios en la variable "productos"
        # Por ejemplo:
        productos = []
        try:
            # Encontrar todos los elementos que contienen los nombres de los productos
            nombres_productos = driver.find_elements(By.XPATH, "//div[@class='descrip_full']")
            # Encontrar todos los elementos que contienen los precios de los productos
            precios_productos = driver.find_elements(By.XPATH, "//span[@class='atg_store_newPrice']")
            # Iterar sobre los nombres y precios de los productos y guardarlos en la lista "productos"
            for nombre, precio in zip(nombres_productos, precios_productos):
                productos.append({'Nombre': nombre.text.strip(), 'Precio': precio.text.strip()})
        except NoSuchElementException:
            print("No se encontraron productos en la página:", url)
        
        # Crear un DataFrame con los productos
        df_productos = pd.DataFrame(productos)
        
        return df_productos
            
    except Exception as e:
        print("Error:", e)
    
    finally:
        # Cerrar el navegador
        driver.quit()


# Llamar a la función para obtener productos y precios
obtener_productos_y_precios('https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-golosinas/_/N-1y5dh9i?Nf=product.endDate%7CGTEQ+1.7130528E12%7C%7Cproduct.startDate%7CLTEQ+1.7130528E12&No=12&Nr=AND%28product.sDisp_200%3A1004%2Cproduct.language%3Aespa%C3%B1ol%2COR%28product.siteId%3ACotoDigital%29%29&Nrpp=437')
```
