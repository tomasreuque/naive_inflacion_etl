```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
url = "https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-panaderia-panaderia-propia/_/N-1s5i92o"
page = urlopen(url)
html = page.read().decode("utf-8")
soup = BeautifulSoup(html, "html.parser")
```

```python
print(soup.get_text())
```

```python
# Busca el elemento con la clase 'descrip_full' que tiene el mismo ID que el nombre del producto
producto = soup.find('div', id='descrip_full_sku00044409')

# Verifica si se encontró el elemento del producto
if producto:
    # Obtiene el siguiente elemento después del producto
    siguiente_elemento = producto.find_next_sibling()

    # Verifica si el siguiente elemento existe y extrae su texto (que sería el precio)
    if siguiente_elemento:
        precio = siguiente_elemento.get_text()
        print("Precio del producto:", precio)
    else:
        print("No se encontró el precio del producto.")
else:
    print("No se encontró el producto.")
```

```python
# Busca el elemento span con la clase 'atg_store_newPrice'
precio_span = soup.find('span', class_='atg_store_newPrice')

# Verifica si se encontró el elemento del precio
if precio_span:
    # Obtiene el texto dentro del elemento span (que debería ser el precio)
    precio = precio_span.get_text(strip=True)
    print("Precio del producto:", precio)
else:
    print("No se encontró el precio del producto.")
```

```python
# Encuentra todos los elementos <span> con la clase 'atg_store_newPrice'
precios = soup.find_all('span', class_='atg_store_newPrice')

# Creamos un diccionario para almacenar los precios únicos y su orden de llegada
precios_unicos = {}

# Itera sobre cada elemento de precio y obtén su texto
for indice, precio_span in enumerate(precios):
    precio = precio_span.get_text(strip=True)
    # Agregamos el precio al diccionario solo si no está presente ya
    if precio not in precios_unicos:
        precios_unicos[precio] = indice

# Imprimimos el diccionario de precios únicos
print("Diccionario de Precios Únicos:")
for precio, indice in precios_unicos.items():
    print("Precio:", precio, "- Orden de llegada:", indice)

```

```python
import requests
from bs4 import BeautifulSoup

# URL base y número de páginas a recorrer
base_url = "https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-panaderia-panaderia-propia/_/N-1s5i92o"
num_paginas = 5  # Define aquí el número de páginas que quieres recorrer

# Estructura de almacenamiento de los productos y precios
productos_precios = {}

# Bucle para recorrer cada página
for pagina in range(1, num_paginas + 1):
    # Construir la URL de la página actual
    url = f"{base_url}?No={pagina * 24}"

    # Realizar la solicitud GET a la página
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Encontrar todos los divs con el ID 'descrip_container_' y los spans con la clase 'atg_store_newPrice'
    productos = soup.find_all('div', id=lambda x: x and x.startswith('descrip_container_sku'))
    precios = soup.find_all('span', class_='atg_store_newPrice')

    # Iterar sobre los productos y precios y almacenarlos en el diccionario
    for producto, precio in zip(productos, precios):
        nombre_producto = producto.get_text(strip=True)
        precio_producto = precio.get_text(strip=True)
        productos_precios[nombre_producto] = precio_producto

# Imprimir el diccionario con todos los productos y precios recopilados
print(productos_precios)
```
